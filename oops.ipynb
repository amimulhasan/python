```python
class Student:
    def __init__(self,name,marks):
      self.name=name
      self.marks=marks

    def get_avg(self):
      sum=0
      for i in self.marks:
        sum+=i
      print(f'hi{self.name} your score is:',sum/len(self.marks))
s1=Student('Amimul',[98,79,89])
s1.get_avg()
s1.name='iranman'
s1.get_avg()
```

```python
class Account:
  def __init__(self,bal,acc):
    self.balance=bal
    self.account_no=acc
  def debit(self,amount):
    self.balance -= amount
    print(f"Taka {amount} was debited from account")
    print('total balance = ',self.get_balance())
  def credit(self,amount):
    self.balance += amount
    print(f' taka {amount} was credited to account')
    print('total balance = ',self.get_balance())
  def get_balance(self):
    return self.balance
acc1=Account(10000,678900)
acc1.debit(1000)
acc1.credit(2000)
```
